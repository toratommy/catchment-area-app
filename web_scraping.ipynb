```python
import pandas as pd # library for data analysis
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents
import pickle
```

```python
def scrape_osm_wiki_amenities(wikiurl):
    response = requests.get(wikiurl)
    # parse data from the html into a beautifulsoup object
    soup = BeautifulSoup(response.text, 'html.parser')
    amenity_table = soup.find('table',{'class':"wikitable"})
    df = pd.read_html(str(amenity_table))
    # convert list to dataframe
    df = pd.DataFrame(df[0])
    df.columns = df.columns.droplevel(1)
    amenity_list = list(df['Value'].str.lower())
    return amenity_list
```

```python
# get the response in the form of html
wikiurl = "https://wiki.openstreetmap.org/wiki/Key:amenity"
```

```python
amenity_list = scrape_osm_wiki_amenities(wikiurl)
```
